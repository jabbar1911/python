```python
# Day 57 - Introduction to Pandas: Series & DataFrames
# ---------------------------------------------------

import pandas as pd
```

```python
# Step 1: Check Pandas version
print("Pandas Version:", pd.__version__)
```

```python
# Step 2: Creating a Series (1D data)
s = pd.Series([10, 20, 30, 40, 50], index=['A', 'B', 'C', 'D', 'E'])
print("\nSeries Example:\n", s)
```

```python
# Step 3: Creating a DataFrame (2D data)
data = {
    'Name': ['Jabbar', 'Saketh', 'Ranga', 'Ankit'],
    'Subject': ['Python', 'Java', 'C', 'AI'],
    'Marks': [90, 85, 88, 92]
}
```

```python
df = pd.DataFrame(data)
print("\nDataFrame Example:\n", df)
```

```python
# Step 4: Accessing DataFrame Information
print("\nShape of DataFrame:", df.shape)
print("Columns:", df.columns)
print("First 2 Rows:\n", df.head(2))
```

```python
# Step 5: Accessing Specific Columns
print("\nNames Column:\n", df['Name'])
```

```python
# Step 6: Describe the numeric data
print("\nStatistical Summary:\n", df.describe())
```
